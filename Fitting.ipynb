## 1D polynomial fitting using numpy
##### http://scipy-lectures.org/intro/numpy/auto_examples/plot_polyfit.html#sphx-glr-intro-numpy-auto-examples-plot-polyfit-py

###### some useful links: https://www.freecodecamp.org/news/data-science-with-python-8-ways-to-do-linear-regression-and-measure-their-speed-b5577d75f8b/

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(12)

# Returns num evenly spaced samples, calculated over the interval [start, stop].
# numpy.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None, axis=0)[source]
x = np.linspace(0, 1, 20)
y = np.cos(x) + 0.3*np.random.rand(20)

# Least squares polynomial fit.
# numpy.polyfit(x, y, deg, rcond=None, full=False, w=None, cov=False)
#    Fit a polynomial p(x) = p[0] * x**deg + ... + p[deg] of degree deg to points (x, y). 
#      Returns a vector of coefficients p that minimises the squared error in the order deg, deg-1, … 0.
#    The solution minimizes the squared error
#    the returns are p : ndarray, shape (deg + 1,) or (deg + 1, K)
#      Polynomial coefficients, highest power first. If y was 2-D, the coefficients for k-th data set are in p[:,k].

# A one-dimensional polynomial class.
# class numpy.poly1d(c_or_r, r=False, variable=None)
#   A convenience class, used to encapsulate “natural” operations on polynomials so that said operations 
#      may take on their customary form in code (see Examples).
polynomial_coefficients, residuals, rank, singular_values, rcond = np.polyfit(x, y, 3, full=True)
p = np.poly1d(polynomial_coefficients)

print('residuals={}'.format(residuals))
print('rank={}'.format(rank))
print('singular_values={}'.format(singular_values))
print('rcond={}\n'.format(rcond))

print('p={}\n'.format(str(p)))
print('roots={}\n'.format(p.r))
print('coefficients={}\n'.format(p.c))
print('order={}\n'.format(p.o))

t = np.linspace(0, 1, 200)
plt.plot(x, y, 'o', t, p(t), '-')

plt.show()
```

## Fitting in Chebyshev basis 

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)

x = np.linspace(-1, 1, 2000)
y = np.cos(x) + 0.3*np.random.rand(2000) + 0.1*np.random.rand(2000) + 0.7*np.random.rand(2000)
p = np.polynomial.Chebyshev.fit(x, y, 90)

t = np.linspace(-1, 1, 200)
plt.plot(x, y, 'r.')
plt.plot(t, p(t), 'k-', lw=3)
plt.show()
```
